```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("./emails.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:, 1:3001]
X
```

```python
Y = df.iloc[:,-1].values
Y
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size = 0.25)
```

```python
# C is used for generalization, it may overfit as C increases
svcModel = SVC(C = 1.0, kernel='rbf', gamma = 'auto') 

svcModel.fit(train_x, train_y)
y_pred = svcModel.predict(test_x)

print("Accuracy Score for SVC: ", accuracy_score(y_pred, test_y))
```

```python
# KNeighborsClassifier

knnModel = KNeighborsClassifier(n_neighbors=7)
knnModel.fit(train_x, train_y)

y_pred = knnModel.predict(test_x)
print(knnModel.score(test_x, test_y))
```
